## EI function plot

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
def ei_fun(x,sigma=1.0):
    return sigma*(x*norm.cdf(x)+norm.pdf(x))
delta = -0.1
sigma = 0.35
beta = delta/sigma
x = np.arange(-3,1,0.01)
plt.plot(x,ei_fun(x), label=r'$\sigma(\beta\Phi(\beta) + \phi(\beta)), (\sigma=1)$')
plt.xlabel(r'$\beta$')
plt.ylabel("EI")
plt.legend()
plt.savefig("EIfun.eps")
```

# Apples and oranges

### Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Data

```python
combos = np.array([[10,0],[9,3],[5,2],[1,4],[0,8]])
combos[1]
```

### Initial Pareto frontier

```python
pareto = np.array([ combos[0], [combos[1,0], combos[0,1]],
                    combos[1], [combos[3,0], combos[1,1]],
                    combos[3], [combos[4,0], combos[3,1]],
                    combos[4]])
# plt.plot(combos[:,1],combos[:,0], '')
plt.plot(combos[0,1],combos[0,0],'bo', label='Combo 1')
plt.plot(combos[1,1],combos[1,0],'go', label='Combo 2')
plt.plot(combos[2,1],combos[2,0],'co', label='Combo 3')
plt.plot(combos[3,1],combos[3,0],'mo', label='Combo 4')
plt.plot(combos[4,1],combos[4,0],'o',color='orange', label='Combo 5')
plt.plot(pareto[:,1],pareto[:,0], 'r-', label='Pareto frontier')
plt.xlabel("Oranges")
plt.ylabel("Apples")
plt.legend()
plt.savefig("pareto_init.eps")
```
